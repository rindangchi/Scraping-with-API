```python
# menggunakan library request untuk membuat koneksi dengan website yang akan di-scraping

import requests

import pandas as pd

from tqdm import tqdm
```

```python
#establih connection to API

resp = requests.get("https://tourney-api.ggwp.id/api/v2/media/getLatestArticle?page=1")

resp.status_code
```

```python
#mengubah response ke dalam bentuk JSON

raw_response = resp.json()

raw_response
```

```python
#mengakses keys dari data yang telah diambil, pada case ini keys-nya adalah [data] [data]

#create dataframe based on json key

quote_data = pd.DataFrame(raw_response["data"]["data"])

quote_data
```

```python
#akan mengambil data dari 10 page saja
#saat menembak API akan diberi jeda 1 detik

import time
```

```python
#create empty list to store value in each page

tmp_data = []

for page in tqdm(range(1,11)):

  #establish connection to API
  resp = requests.get(f"https://tourney-api.ggwp.id/api/v2/media/getLatestArticle?page={page}")

  #convert response to json
  raw_response = resp.json()

  #get data based on key
  get_data = raw_response["data"]["data"]

  #get data into the list
  tmp_data.extend(get_data)

  #give interval 1 second
  time.sleep(1)




```

```python
#ubah list ke dalam bentuk dataframe

final_data = pd.DataFrame(tmp_data)

final_data
```
